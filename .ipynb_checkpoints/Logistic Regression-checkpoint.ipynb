```python
def LogisticRegression(X,y):
        
    X=np.insert(X,0,1,axis=1)
    W=np.ones(X.shape[1])
    lr=0.5
        
    for i in range(5000):
            
        yhat= sigmoid(np.dot(X,W))
        W= W + lr * (np.dot((y - yhat),X)/X.shape[0])
        
    return W[0],W[1:]
        
```

```python
def sigmoid(z):
    return 1/(1+ np.exp(-z))
```
